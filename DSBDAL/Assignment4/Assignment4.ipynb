```python
# Relational Plot Functions
# These functions map data variables to visual properties like position, color, and size.
# - relplot(): Figure-level interface combining scatterplot or lineplot with faceting. Use kind="scatter" (default) or "line".
# - scatterplot(): Axes-level scatter plot showing individual data points.
# - lineplot(): Axes-level line plot showing trends, with confidence intervals by default.
```

```python
# Distribution Plot Functions
# These handle single or paired variables with semantic mappings.
# - displot(): Figure-level interface; supports kind="hist" (default), "kde", "ecdf".
# - histplot(): Axes-level histogram with optional KDE overlay.
# - kdeplot(): Axes-level kernel density estimate curve.
# - ecdfplot(): Axes-level empirical cumulative distribution function.
# - rugplot(): Axes-level rug plot showing data point positions.
```

```python
# Figure-level: catplot(kind=...) wraps axes-level plots with faceting.
# Three families:
# - Scatterplots: stripplot(), swarmplot()
# - Distribution: boxplot(), violinplot(), boxenplot()
# - Statistical: barplot(), pointplot()
```

```python
# replot()
# data: Dataset (DataFrame preferred).
# x, y: Variables for x/y axes.
# hue: Variable for color encoding (categorical/numeric).
# size: Variable for point/line size.
# style: Variable for markers/dashes.
# row, col: Variables for subplot faceting.
# kind: "scatter" or "line".
# height: Facet height (inches); aspect: Width/height ratio.
# palette: Color scheme (e.g., "Set1").
# legend: "auto", "brief", "full", or False.


df.dtypes
sns.relplot(data=df, x="Age", y="Fare", hue="Pclass", kind="scatter")
sns.relplot(data=df, x="Age", y="Fare", hue="Pclass", col="Sex", kind="line",linestyle='--',)
sns.relplot(data=df, x="Age", y="Pclass", hue="Survived", col="Sex")
```

```python
# scatterplot()
# Axes-level function for scatter plots. Shares relational params with relplot:
# - data, x, y, hue, size, style: Same as relplot.
# - alpha: Point transparency (0-1).
# - linewidth: Edge thickness.
# - markers: Specific marker shapes (bool or list).
# - edgecolor: Edge color.

# sns.scatterplot(df,x='Age',y='Fare',hue='Survived')
# sns.scatterplot(df,x='Age',y='Fare',alpha=0)
# sns.scatterplot(df,x='Age',y='Fare',edgecolor='r',linewidth=1.5)
# sns.scatterplot(df,x='Age',y='Fare',  color='r', marker='s')
markers=['0','s','^','*']
# sns.scatterplot(df,x='Age',y='Pclass',hue='Sex',style='Sex', markers=True)
sns.scatterplot(df,x='Age',y='Pclass',hue='Survived',style='Survived', markers=['^','*'])
```

```python
# lineplot()
# Axes-level function for lines with error bands. Key params
# - data, x, y, hue, size, style: Same as above.
# - estimator: Aggregation ("mean" default); None for all points.
# - sort: Sort x values (bool).
# - dashes: Use dashes for styles (bool).Set dashes=True (default)
# - marker: Marker symbol ("o", "s", etc.).
# - errorbar: Error type ("ci" for intervals, "sd", or None).
        # "ci": Bootstrap confidence interval (default 95%, adjustable via errorbar=("ci", 90)).
        # "sd": Standard deviation band.
        # "pi": Prediction interval.
        # None or "none": No error bands.
        # (percentile_low, percentile_high): Custom percentiles, e.g., (10, 90).

# sns.lineplot(df,x='Age',y='Pclass')
# sns.lineplot(df,x='Age',y='Fare',sort=False)

# sns.lineplot(df,x='Age',y='Fare')
# sns.lineplot(df,x='Age',y='Fare',marker='s')
# sns.lineplot(df, x="Age", y="Fare", hue="Pclass", style="Pclass")
# sns.lineplot(df, x="Age", y="Fare", hue="Pclass", style="Pclass", dashes=False)
# sns.lineplot(df, x="Pclass", y="Fare", estimator="median", hue="Sex")
# sns.lineplot(df, x="Age", y="Fare", hue="Sex",errorbar=('ci',90))
sns.lineplot(df, x="Age", y="Fare", hue="Sex",errorbar=('sd',90))
# sns.lineplot(df, x="Age", y="Fare", hue="Sex",errorbar=('pi',90))
```

```python
# displot Parameters
# Figure-level function for distributions with faceting. Key parameters:
# - data: Dataset (DataFrame or array).
# - x (or y for bivariate): Variable(s) to plot.
# - hue: Grouping for colors/fills.
# - kind: "hist", "kde", "ecdf".
# - multiple: "layer", "dodge", "stack", "fill" for hue handling.
# - bins: Histogram bin count or edges.
# - kde: Add KDE curve (bool or kwargs).
# - fill: Shade under KDE (bool).
# - col, row: Facet variables.
# - height, aspect: Facet sizing.

# sns.displot(data=df, x="Age",y='Fare', hue="Pclass", kind="ecdf")
# sns.displot(data=df, x="Age", hue="Pclass", kind="kde")
# sns.displot(data=df, x="Age", hue="Pclass", kind="hist")

# sns.displot(data=df, x="Age", y='Fare', hue="Pclass", kind="hist")

# sns.displot(data=df, x="Age", hue="Sex", multiple="dodge")
# sns.displot(data=df, x="Age", hue="Sex", multiple="layer")
# sns.displot(data=df, x="Age", hue="Sex", multiple="stack")
# sns.displot(data=df, x="Age", hue="Sex", multiple="fill")

# sns.displot(data=df, x="Age",kde=True)
# sns.displot(data=df, x="Fare", kind="kde", fill=True)
# sns.displot(data=df, x="Age", col="Sex")
# sns.displot(data=df, x="Fare", row="Pclass")
sns.displot(data=df, x="Age", hue="Pclass", col="Survived", height=4, aspect=1.5)
# sns.displot(data=df, x="Fare", log_scale=True)





```

```python
# histplot Parameters
# Axes-level histograms. Shares params with displot:
# - data, x, y, hue, bins, kde, multiple: Same as above.
# - binwidth: Fixed bin width.
# - binrange: (min, max) bin limits.
# - cumulative: Cumulative histogram (bool).
# - stat: "count", "probability", "percent", "density".


# sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="count",cumulative=True)
# sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="count",binrange=(10,500))
# sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="count",binrange=(100,500))
# sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="density")
sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="percent")
# sns.histplot(data=df, x="Age", hue="Sex", kde=True, multiple="dodge",stat="probability")

```

```python
# kdeplot Parameters
# Smooth density curves without bins:
# - data, x, y, hue, fill: Same as above.
# - bw_adjust: Bandwidth adjustment factor.
# -levels: Contour levels for bivariate.
# - thresh: Minimum density threshold (bivariate).
# - gridsize: Evaluation grid resolution.


# sns.kdeplot(df,x='Age',hue='Pclass')
# sns.kdeplot(data=df, x="Age", hue="Pclass",levels=5,  multiple="stack",fill=True)
# sns.kdeplot(data=df, x="Age", y="Fare",levels=5)
# sns.kdeplot(data=df, x="Age", y="Fare",levels=10,  fill=True)
sns.kdeplot(data=df, x="Age", y="Fare",levels=10, thresh=0.1, fill=True)    #Clips low-density regions, focusing central mass
```

```python
# ecdfplot:
# - data, x, y, hue: Same semantics.
# - stat: "empirical" or "proportion".
# rugplot:
# - data, x, y, hue: Data positions as ticks.
# - height: Tick height (0-1).
# - expand: Axis padding.


# sns.ecdfplot(df,x='Age',hue='Sex')


# sns.scatterplot(data=df, x="Age", y="Fare")
sns.rugplot(data=df, x="Age", y="Fare", hue="Survived")



```

```python
# pairplot Parameters:
# - data: DataFrame.
# - hue: Color grouping variable.
# - kind: "scatter" (default), "kde", "hist", "reg".
# - diag_kind: "auto", "hist", "kde".
# - corner: Hide upper triangle.
# - plot_kws: Dict of kwargs for off-diagonal plots.


sns.pairplot(df, hue="Pclass", diag_kind="kde")
```

```python
# jointplot Parameters:
# - data, x, y: Variables.
# - hue: Color grouping.
# - kind: "scatter", "kde", "hist", "hex", "reg", "resid".
# - height: Figure size (inches).
# - marginal_kws, scatter_kws: Subplot kwargs.


sns.jointplot(data=df, x="Age", y="Fare")
```
