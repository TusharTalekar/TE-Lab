```python
# Convert varirable to Proper Data Type
#   1. Convert Sruvived and Pclass to Cate# Convert Categorial Variables into Quantitative Variables
#  1. Sexgorial
#   2. Convert Sex and Embarked to Categorial
```

```python
df_first = df.copy()
df_first['Survived'] = df_first['Survived'].astype('category')
df_first['Pclass'] = df_first['Pclass'].astype('category')
df_first.head()
df_first.dtypes
```

```python
# 2. Convert Sex and Embarked to Categorial
df_second = df.copy()
df_second['Sex'] = df_second['Sex'].astype('category')
df_second['Embarked'] = df_second['Embarked'].astype('category')
df_second.head()
df_second.dtypes
```

```python
# Fill the Age in Place with mean_age
df_third = df.copy()
mean_age = df_third['Age'].mean()
df_third['Age']=df_third['Age'].fillna(mean_age)
df_third
```

```python
# One Hot Encoding for Sex and Embarked
df_one_encoded = pd.get_dummies(df_third, columns=['Sex', 'Embarked'], drop_first=True)
df_one_encoded.head()
```
