```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.datasets import load_iris
```

```python
df=df.dropna(subset=['College','Salary'])
cols_with_null_cleaned = df.isnull().any()
print(cols_with_null_cleaned)
# Count of data being null
df.isnull().sum()
```

```python
# Mode of College Columns
# Here as it is String
# So mode() simply returns the most frequently occurring value(s) in the column 
# regardless of whether the data is numeric or text.
print(df["College"].value_counts())
df["College"].mode()
```

```python
df.groupby('Position')['Age'].mean()
```

```python
# Standard deviation
df['Age'].std()
```

```python
# Group by Position and calculate summary statistics for Age
summary = df.groupby('Position')['Age'].agg(
    mean='mean',
    median='median',
    minimum='min',
    maximum='max',
    std_deviation='std'
)
print(summary)

# Group by postion and average height
summary = df.groupby('Position')['Weight'].agg(
    mean='mean',
    median='median',
    minimum='min',
    maximum='max',
    std_deviation='std'
)
print(summary)
```

```python
# Display statistics for each species

for sp in df['Team'].unique():
    print(f"\nStatistics for {sp}")
    print(df[df['Team'] == sp].describe())

# for sp in df['College'].unique():
#     print(f"\nStatistics for {sp}")
#     print(df[df['College'] == sp].describe())
```
