```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load CSV into DataFrame 
df = pd.read_csv("Iris.csv")
```

```python
df.shape
```

```python
df.head()
df.dtypes
```

```python
df['Species'].value_counts()
```

```python
summary = df.groupby('Species')['SepalLengthCm'].agg(
    mean='mean',
    median='median',
    minimum='min',
    maximum='max',
    std_deviation='std'
)
print(summary)
```

```python
for species in df["Species"].unique():
        print(f"\nStatistical details for species: {species}\n")
        stats = (
            df[df["Species"] == species]
            .drop(columns=["Species"])
            .describe()
            .transpose()
        )
        
        display(stats)
```

```python
std_df = df.groupby("Species").std()
std_df
```

```python
std_df.plot( figsize=(8,5))
plt.show()
```
