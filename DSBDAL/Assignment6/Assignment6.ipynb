```python
fig,axes=plt.subplots(2,2,figsize=(10,8))

sns.histplot(df,x='SepalLengthCm',hue='Species',multiple='layer',ax=axes[0,1])
sns.histplot(df,x='SepalWidthCm',hue='Species',multiple='layer',ax=axes[1,1])
sns.histplot(df,x='PetalLengthCm',hue='Species',multiple='layer',ax=axes[1,0])
sns.histplot(df,x='PetalWidthCm',hue='Species',multiple='layer',ax=axes[0,0])

plt.tight_layout()
plt.show()

# Sepal features
# Heavy overlap between species
# Broad, almost normal-like distributions
# Hard to classify using sepal data alone
# - Weak predictive power

# Petal features
# Clear multi-modal distributions
# Minimal overlap
# Distinct peaks for each species
# - Strong predictive power
```

```python
fig,axes=plt.subplots(2,2,figsize=(9,7))

sns.boxplot(df,x='Species',y='SepalLengthCm',hue='Species',ax=axes[0,0])
sns.boxplot(df,x='Species',y='PetalLengthCm',hue='Species',ax=axes[0,1])
sns.boxplot(df,x='Species',y='PetalWidthCm',hue='Species',ax=axes[1,0])
sns.boxplot(df,x='Species',y='SepalLengthCm',hue='Species',ax=axes[1,1])

plt.tight_layout()
plt.show()

# Sepal Length
# Setosa
    # Smallest range
    # Very tight distribution
    # Almost no outliers
# Versicolor & Virginica
    # Larger spread
    # Heavy overlap
    # A few points near whisker edges → mild outliers
# 📌 Conclusion: Sepal length does not separate species well.

# Petal Length
# Setosa
    # Very small values
    # Completely separated from others
    # No outliers
# Versicolor
    # Moderate spread
# Virginica
    # Largest values
    # Slight outliers on the upper side
# Conclusion: Petal length is excellent for classification.

# Petal Width
    # Same pattern as petal length, but even cleaner
    # Almost zero overlap
    # Clear whisker separation
    # A few mild high-end outliers for Virginica
# Conclusion: Best feature for species discrimination.
```
